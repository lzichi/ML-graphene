## Functions

```python
def tohex(array):
    return np.where(array == True, 16738740, 000000)
```

```python
def mark_image(Image, image):
    
    #Make a copy of the image to use for all the next steps
    img = image.copy()
    
    # Get the color of the click 
    imarray = img
    
    # Draw a cricle around the user mouse click on the flake 
    def draw_circle(event,x,y,flags,param):
        '''Draw a circle if you click once on the house and print the color of the image at that point'''
        global mouseX,mouseY, R, G, B
        #If single click draw a cricle centered on the click in blue
        if event == 1:
            cv2.circle(img,(x,y),20,(255,0,0),thickness=10)
            #print("Color: (R, G, B): (", imarray[y,x,0], ",", imarray[y,x,1], ",", imarray[y,x,2], ")")
            
            mouseX, mouseY, R, G, B = x, y, imarray[y,x,0], imarray[y,x,1], imarray[y,x,2]
        
    # Create a window for the image and link it to the mouse 
    cv2.startWindowThread()
    cv2.namedWindow('image')
    cv2.setMouseCallback('image', draw_circle)

    # Create an infinite loop that ends when you press the space bar
    while(1):
        cv2.imshow('image', cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        # end when the q button is pressed 
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
        
    # Destroy the window
    cv2.waitKey(1)
    cv2.destroyAllWindows()
    cv2.waitKey(1)

    
    # find the corresponding pixels 
    flake = (R, G, B)

    return mouseX, mouseY, flake 
```

```python
def mask_image(path, isDisplay, ncolors = 256):
    '''Take in an image and return image with flake in white and 
    everything else in black. Display original and masked in specified.'''

    org_img = Image.open(path)
    org_arr = np.asarray(org_img)
    
    # create masked area for orginal image 
    org_img = org_img.quantize(colors = ncolors, dither=None)
    org_img = org_img.convert('RGB')
    org_img_arr = np.asarray(org_img)
    mouseX, mouseY, flake = mark_image(org_img, org_img_arr) 
    
    a = np.all(org_img_arr == flake, axis = -1)
    a = a.reshape((org_img_arr.shape[0], org_img_arr.shape[1]))
    arr = tohex(a)
       
    if(isDisplay):
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
        ax1.imshow(org_img)
        ax1.axis("off")
        #ax1.set_title("Original Image")
       
        ax2.imshow(arr, cmap = 'gray') 
        ax2.axis("off")
        #ax2.set_title("Masked Image")
    plt.show()
    fig.savefig('bad_CAM.png')

    return arr
```

```python
## Code to Read in Images from Dropbox and put them in a skeleton dataFrame with their path, ##
##  name, material and magnification ##
rootdir = r'/Users/laurazichi/Dropbox/Summer 2020 Machine Learning Project/Pictures/'
#create empty arrays for the parameters of the DataFrame
images_path = []
images_name = []
materials = []
magnification = []

#for loop to go through folders of images and add a row for each data instance
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        images_name.append(file)
        images_path.append(os.path.join(subdir,file))
        mag = file.split('x')[0].split('-')[-1]
        magnification.append(mag)
        mat = os.path.join(subdir,file).split("x-")[-1].split(".")[0]
        materials.append(mat)

            
data = {'photo_name': images_name,
       'path': images_path,
        'magnification': magnification,
       'materials': materials}

df = pd.DataFrame.from_dict(data)

## Create Dataframe of only MoSe2 Images ##
df = df[~df.photo_name.str.contains('Screen Shot')]
#df = df[~df.photo_name.str.contains('cropped')]
df = df[~df.magnification.str.contains('50')]
df_MoSe2 = df[df.materials.str.contains('MoSe2')]
df_MoSe2.reset_index(drop=True, inplace=True)
```
