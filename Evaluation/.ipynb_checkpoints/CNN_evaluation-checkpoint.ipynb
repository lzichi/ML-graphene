## Functions

### Helpers

```python
def tohex(array):
    return np.where(array.reshape(224, 224) == True, 16738740, 0)
```

```python
def df_paths(rootdir):
    ''' Code to Read in Images from Dropbox and put them in a skeleton dataFrame with their path,
        name, material and magnification '''
    
    #create empty arrays for the parameters of the DataFrame
    images_path = []
    images_name = []
    materials = []
    magnification = []

    #for loop to go through folders of images and add a row for each data instance
    for subdir, dirs, files in os.walk(rootdir):
        for file in files:
            images_name.append(file)
            images_path.append(os.path.join(subdir,file))
            mag = file.split('x')[0].split('-')[-1]
            magnification.append(mag)
            mat = os.path.join(subdir,file).split("x-")[-1].split(".")[0]
            materials.append(mat)


    data = {'photo_name': images_name,
           'path': images_path,
            'magnification': magnification,
           'materials': materials}

    df = pd.DataFrame.from_dict(data)
    
    return df
```

```python
def reverse(path_cam, path_org, alpha = 0.4):

    cam = np.asarray(Image.open(path_cam).resize((224, 224)))
    cam_arr = np.asarray(cam, dtype = 'int')

    img = Image.open(path_org)
    img = img.resize((cam_arr.shape[0], cam_arr.shape[1]))
    img_arr = np.asarray(img, dtype = 'int')
    
    
    sub_arr = cam_arr - img_arr
    sub_arr_flat = sub_arr.flatten()
 
    sub_arr = np.where(sub_arr < 0, 0, sub_arr)
    #sub_arr = np.where(sub_arr > 255, 255, sub_arr)
    #jet_heatmap = cam_arr
    jet_heatmap = sub_arr

    jet_heatmap = jet_heatmap/256
    
    # Use jet colormap to colorize heatmap
    jet = cm.get_cmap("jet")
    
    jet_heatmap = np.round(jet_heatmap, decimals = 6)
    
    # Use RGB values of the colormap
    jet_colors = jet(np.arange(256))[:, :3] 
    heatmap = np.zeros((np.shape(jet_heatmap)[0], np.shape(jet_heatmap)[1]))
    
    for ind_row, row in enumerate(jet_heatmap):
        for ind_col, color in enumerate(row):
            sub = jet_colors - color
            sub = sub * sub
            index = np.argmin(np.sum(sub, axis = 1))
            
            heatmap[ind_row][ind_col] = index
     
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    heatmap = np.array(heatmap)

    
    fig, ax = plt.subplots(1, 4, figsize=(20,10))
    ax[0].imshow(cam_arr)
    ax[1].imshow(img)
    ax[2].imshow(sub_arr)
    ax[3].matshow(heatmap)

    return heatmap
```

```python
def mark_image(Image, image):
    
    #Make a copy of the image to use for all the next steps
    img = image.copy()
    
    # Get the color of the click 
    imarray = img
    
    # Draw a cricle around the user mouse click on the flake 
    def draw_circle(event,x,y,flags,param):
        '''Draw a circle if you click once on the house and print the color of the image at that point'''
        global mouseX,mouseY, R, G, B
        #If single click draw a cricle centered on the click in blue
        if event == 1:
            cv2.circle(img,(x,y),20,(255,0,0),thickness=10)
            #print("Color: (R, G, B): (", imarray[y,x,0], ",", imarray[y,x,1], ",", imarray[y,x,2], ")")
            
            mouseX, mouseY, R, G, B = x, y, imarray[y,x,0], imarray[y,x,1], imarray[y,x,2]
        
    # Create a window for the image and link it to the mouse 
    cv2.startWindowThread()
    cv2.namedWindow('image')
    cv2.setMouseCallback('image', draw_circle)

    # Create an infinite loop that ends when you press the space bar
    while(1):
        cv2.imshow('image', cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        # end when the q button is pressed 
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
        
    # Destroy the window
    cv2.waitKey(1)
    cv2.destroyAllWindows()
    cv2.waitKey(1)

    
    # find the corresponding pixels 
    flake = (R, G, B)
    arr = np.asarray([*np.array(Image.getcolors(), dtype = object)[:,1]])
    arr2 = np.asarray([*np.array(Image.getcolors(), dtype = object)[:,0]])
    index = np.where(np.all(arr == flake, axis = 1) == True)[0][0]
    area = arr2[index]
    
    #print('flake area: ', area)
    
    # return area of true where equal to cliked color and false otherwise
    masked_arr = np.all(image == flake, axis = -1).flatten()
   
    return mouseX, mouseY, R, G, B, area, masked_arr
    
```

```python
def find_overlap_new(path_cam, path_org, df, isShow, org_quant = 150):
    '''Take in a cam and original image and add cam_path, org_path, flake_area, cam_area, overlap, cam_color 
        to dataframe '''
    
    # open cam image and convert to array
    cam_im = Image.open(path_cam)
    cam_arr = np.asarray(cam_im)

    # open, reshape original image to same dimensions as cam
    org_im = Image.open(path_org)
    org_img = org_im.resize((cam_arr.shape[0], cam_arr.shape[1]))
    org_arr = np.asarray(org_img)

    if(isShow):
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
        ax1.imshow(cam_arr)
        ax1.axis("off")
        ax1.set_title("Grad-CAM", fontsize = 20, **hfont)
        ax2.imshow(org_arr) 
        ax2.axis("off")
        ax2.set_title("Original", fontsize = 20, **hfont)
    plt.show()
    #fig.savefig('good_CAM.png')

    # create masked area for orginal image 
    org_img = org_img.quantize(colors = org_quant, dither=None)
    org_img = org_img.convert('RGB')
    org_img_arr = np.asarray(org_img)
    mouseX, mouseY, R, G, B, flake_area, org_masked_arr = mark_image(org_img, org_img_arr)
    #print('flake area: ', np.shape(np.where(org_masked_arr == True)))

    heatmap = reverse(path_cam, path_org)
    heatmap = heatmap.flatten()
    
    # combine arrays and find overlap of true region 
    combined = np.where(org_masked_arr == True, heatmap, 0)
    overlap_area = np.sum(combined)
    
    if(isShow):
        
        arr3 = combined.reshape((224, 224))
        arr4 = tohex(org_masked_arr)

        fig, axs = plt.subplots(1, 2, figsize=(15,5))
        for ax, arr in zip (axs, np.array([arr3, arr4])):
            ax.imshow(arr)
            ax.axis("off")
    plt.show()
    
    print('Are you happy? Y/N')
    user_input = input()
    if(user_input == 'Y'):
        # add to end of dataframe 
        df.loc[len(df.index)] = [path_cam, path_org, flake_area, overlap_area]
        display(df)
        return False
    else:
        return True
```
